```python
covid_df.loc[(covid_df['Sexo'] == 'f'),['Sexo']] = 'F'
covid_df.loc[(covid_df['Sexo'] == 'm'),['Sexo']] = 'M'
covid_df.loc[(covid_df['Estado'] == 'Fallecido'),['Estado']] = 'FALLECIDO'
covid_df.loc[(covid_df['Estado'] == 'Grave'),['Estado']] = 'GRAVE'
covid_df.loc[(covid_df['Estado'] == 'leve'),['Estado']] = 'LEVE'
covid_df.loc[(covid_df['Estado'] == 'Leve'),['Estado']] = 'LEVE'
covid_df.loc[(covid_df['Estado'] == 'Moderado'),['Estado']] = 'MODERADO'
```

```python
df_sexo_estado = covid_df.groupby(["Sexo","Estado"])['ID de caso'].count()
df_sexo_estado
```

```python
df = covid_df.groupby(["Sexo","Estado"])['Edad'].mean()
df
```

```python
df = covid_df.groupby(["Sexo","Estado"])['Edad'].max()
df
```

```python
df_fallecidos = covid_df[(covid_df['Estado']=='FALLECIDO')]
df_fallecidos.shape
```

```python
df_fallecidos_mean =  df_fallecidos.groupby('Estado')['Edad'].transform('mean')
display(
    df_fallecidos_mean.head(),
    df_fallecidos_mean.shape
)
```

```python
df_fallecidos['Edad promedio'] = df_fallecidos_mean
df_fallecidos.head()
```

```python
df_fallecidos_pordebajo = df_fallecidos[df_fallecidos['Edad'] < df_fallecidos['Edad promedio']]
display(df_fallecidos_pordebajo.head(),df_fallecidos_pordebajo.shape)
```
